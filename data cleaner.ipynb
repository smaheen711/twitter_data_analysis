```python
# df_main = pd.read_csv('cleaned_scrapped_tweets_ru.csv')
df_main = pd.read_csv('cleaned_translated_5500.csv',low_memory=False)

df_main.head()
```

```python
translator = Translator()

for i in range(5501,len(df_main['Translated Text'])):
    try:
        if df_main['Translated Text'][i] == 0 or'0':
            text = df_main['Cleaned Text'][i]
            translated_text = translator.translate(text=text, dest='en').text
            print(f"Current Number: {i}")
            print(f"Translated Text: {translated_text}")
            df_main['Translated Text'][i] = translated_text
            time.sleep(3)
    except:
        continue
    if i%500 == 0:
        df_main.to_csv(f"cleaned_translated_{i}.csv")
        
```
