```python
import pandas as pd
import numpy as np
from googletrans import Translator
import time
```

```python
df_main = pd.read_csv('cleaned_scrapped_tweets_ru.csv')

df_main.head()
```

```python
df_main.describe()
```

```python
df_main.drop_duplicates('Cleaned Text', inplace=True)
df_main.reset_index(inplace=True)
```

```python
df_main.describe()
```

```python
df_main['Translated Text'] = 0
```

```python
translator = Translator()

for i in range(len(df_main['Translated Text'])):
    try:
        if df_main['Translated Text'][i] == 0:
            text = df_main['Cleaned Text'][i]
            translated_text = translator.translate(text=text, dest='en').text
            print(f"Current Number: {i}")
            print(f"Translated Text: {translated_text}")
            df_main['Translated Text'][i] = translated_text
            time.sleep(3)
    except:
        continue
        
        
```
